# Task 4 - Recommendation System

## Objective
Build a Content-Based Recommendation System using CountVectorizer and Cosine Similarity.

## Technologies Used
- Python
- Pandas
- Scikit-learn
- Jupyter Notebook

Import Libraries

```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.metrics.pairwise import cosine_similarity

print("Libraries Loaded Successfully")
```

Create Dataset

```python
movies = {
    "Movie": [
        "Avatar",
        "Titanic",
        "Avengers",
        "Iron Man",
        "The Lion King",
        "Black Panther",
        "Spider-Man"
    ],

    "Genre": [
        "Action Adventure",
        "Romance Drama",
        "Action Superhero",
        "Action Superhero",
        "Animation Family",
        "Action Superhero",
        "Action Adventure"
    ]
}

df = pd.DataFrame(movies)

df
```

Vectorization

```python
vectorizer = CountVectorizer()

vectors = vectorizer.fit_transform(df["Genre"])

print(vectors.shape)
```

Cosine Similarity

```python
similarity = cosine_similarity(vectors)

print(similarity)
```

Recommendation Function

```python
def recommend(movie_name):

    index = df[df["Movie"] == movie_name].index[0]

    distances = similarity[index]

    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )

    print("Recommended Movies:\n")

    for movie in movie_list[1:4]:
        print(df.iloc[movie[0]].Movie)
```

Test

```python
recommend("Avengers")
```
